```python
import pandas as pd; pd.options.mode.chained_assignment = None
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.cluster import KMeans
import datetime as dt
import xgboost as xgb
import numpy as np
import random
import math

random.seed(7662)
np.random.seed(7662)

def MultiVarianceImputer(df):
    mvi_range = range(len(df))
    mvi_mean = df.median(axis=0)
    for mvi_c in df.columns:
        mvi_df = df[df[mvi_c].notnull()]
        mvi_y = mvi_df[mvi_c].values
        mvi_df = mvi_df[mvi_df.columns.difference([mvi_c])].fillna(mvi_mean).fillna(0)
        if len(mvi_df)>1 and len(mvi_df)<(len(df)-1):
            etr = ExtraTreesRegressor(n_estimators=100, max_depth=4, n_jobs=-1, random_state=1)
            etr.fit(mvi_df, mvi_y)
            mvi_pred = etr.predict(df[df.columns.difference([mvi_c])].fillna(mvi_mean).fillna(0))
            df[mvi_c] = [mvi_pred[i] if pd.isnull(df[mvi_c][i]) else df[mvi_c][i] for i in mvi_range]
        #print(mvi_c)
    df = df.fillna(0)
    return df

def get_features(df, mvi=False, cut=0):
    df_cut = df[:cut]
    df = df[cut:]
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    df["timestamp_year"] = df["timestamp"].dt.year
    df["timestamp_month"] = df["timestamp"].dt.month
    df["timestamp_day"] = df["timestamp"].dt.day
    df['timestamp_hour'] = df["timestamp"].dt.hour
    df['timestamp_weekday'] = df['timestamp'].dt.weekday
    df['timestamp_week'] = df['timestamp'].dt.week
    df['timestamp_quarter'] = df['timestamp'].dt.quarter
    df['timestamp_weekend'] = ((df['timestamp_weekday'] == 5) & (df['timestamp_weekday'] == 6))
    df['timestamp_wd'] = ((df['timestamp_weekday'] != 5) & (df['timestamp_weekday'] != 6))
    df['timestamp'] = df['timestamp'].map(lambda x: float((x - dt.datetime(1899, 12, 30)).days) + (float((x - dt.datetime(1899, 12, 30)).seconds) / 86400))
    df = pd.concat((df_cut, df), axis=0, ignore_index=True).reset_index(drop=True)
    if mvi==True:
        df = MultiVarianceImputer(df)
    return df

def rmsle(y, pred):
    assert len(y) == len(pred)
    return np.sqrt(np.mean(np.power(np.log1p(y)-np.log1p(pred), 2)))

train = pd.read_csv('train.csv', parse_dates=['timestamp']).tail(100) #change this
test = pd.read_csv('test.csv', parse_dates=['timestamp']).tail(100) #change this
macro = pd.read_csv('macro.csv', parse_dates=['timestamp'])
train = pd.merge(train, macro, how='left', on='timestamp')
test = pd.merge(test, macro, how='left', on='timestamp')
print(len(train),len(test),len(train.columns))

col = [c for c in train.columns if c not in ['id','price_doc']]
colt = [c for c in train.columns if c not in ['price_doc']]

cat = [train.columns[c] for c in range(len(train.columns)) if train.dtypes[c] == 'O']
lbl = LabelEncoder()
for c in cat:
    lbl.fit(list(train[c].fillna('').values)+list(test[c].fillna('').values))
    train[c] = lbl.transform(list(train[c].fillna('').values))
    test[c] = lbl.transform(list(test[c].fillna('').values))
    print(c, len(lbl.classes_))

train = get_features(train, True)
test = pd.concat((train, test), axis=0, ignore_index=True).reset_index(drop=True)
test = test[colt]
test = get_features(test, True, len(train))[len(train):]
col = [c for c in train.columns if c not in ['id','price_doc']]

print(len(train),len(test),len(train.columns),len(col))
```

```python
k = KMeans(n_clusters=10, max_iter=10, random_state=1, n_jobs=-1)
k.fit(train[col])
train['kmeans1'] = k.labels_
kpred = k.predict(test[col])
test['kmeans1'] = kpred

k.fit(train[['full_sq','num_room','life_sq']])
train['kmeans2'] = k.labels_
kpred = k.predict(test[['full_sq','num_room','life_sq']])
test['kmeans2'] = kpred
col += ['kmeans1','kmeans2']

train.timestamp_wd = train.timestamp_wd.astype(int)
train.timestamp_weekend = train.timestamp_weekend.astype(int)
train.kmeans1 = train.kmeans1.astype(int)
train.kmeans2 = train.kmeans2.astype(int)

test.timestamp_wd = test.timestamp_wd.astype(int)
test.timestamp_weekend = test.timestamp_weekend.astype(int)
test.kmeans1 = test.kmeans1.astype(int)
test.kmeans2 = test.kmeans2.astype(int)
```

```python
from sklearn import preprocessing
rs = preprocessing.RobustScaler(with_centering=False, with_scaling=False, quantile_range=(25.0, 75.0), copy=True)

etr = ExtraTreesRegressor(n_estimators=1000, max_depth=20, n_jobs=-1, random_state=1)
etr.fit(rs.fit_transform(train[col]), train['price_doc']); print('etr fit...')
print(rmsle(train['price_doc'], etr.predict(rs.transform(train[col]))))

pred =etr.predict(rs.transform(test[col])); print('etr predict...')
test['price_doc'] = pred
test[['id','price_doc']].to_csv('submission.csv', index=False)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

imp = etr.feature_importances_
feat_imp = [[imp[i], col[i]] for i in range(len(imp))]
_ = pd.DataFrame(feat_imp, columns=['importance','column']).sort_values(['importance','column'], ascending=[False, False])[:20].plot(kind='barh', x='column', y='importance')
```

```python
import xgboost as xgb

xgb_params = {
    'eta': 0.02,
    'max_depth': 20,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1
}

dtrain = xgb.DMatrix(train[col], train['price_doc'])
dtest = xgb.DMatrix(test[col])

cv_output = xgb.cv(xgb_params, dtrain, num_boost_round=1000, early_stopping_rounds=20, verbose_eval=50, show_stdv=False)
cv_output[['train-rmse-mean', 'test-rmse-mean']].plot()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

num_boost_rounds = len(cv_output)
model = xgb.train(dict(xgb_params, silent=0), dtrain, num_boost_round= num_boost_rounds)

plt.rcParams['figure.figsize'] = (8.0, 15.0)
xgb.plot_importance(model); plt.show()
```

```python
pred = model.predict(dtest)
test['price_doc'] = pred
test[['id','price_doc']].to_csv('submissions/xgb5.csv', index=False)
```
