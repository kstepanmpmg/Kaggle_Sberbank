```python
def rmsle(preds, labels, XGB=False):
#     labels = dtrain.get_label()
    if not XGB:
#         pass
#     else:
        assert len(preds) == len(labels)
        preds = preds.tolist()

    labels = labels.tolist()
    terms_to_sum = [(math.log(labels[i] + 1) - math.log(max(0, preds[i]) + 1)) ** 2.0 for i, pred in enumerate(labels)]
    return 'rmsle', (sum(terms_to_sum) * (1.0 / len(preds))) ** 0.5
```

# XGBoost

```python
# def rmsle_XGB(preds, labels):
#     preds = list(preds)
#     labels = labels.get_label()
#     assert len(preds) == len(labels)
#     terms_to_sum = [(math.log(labels[i] + 1) - math.log(max(0, preds[i]) + 1)) ** 2.0 for i, pred in enumerate(labels)]
#     return 'rmsle', (sum(terms_to_sum) * (1.0 / len(preds))) ** 0.5

def rmsle_XGB(preds, dtrain):
    labels = dtrain.get_label()
    assert len(preds) == len(labels)
    labels = labels.tolist()
    preds = preds.tolist()
    terms_to_sum = [(math.log(labels[i] + 1) - math.log(max(0, preds[i]) + 1)) ** 2.0 for i, pred in enumerate(labels)]
    return 'rmsle', (sum(terms_to_sum) * (1.0 / len(preds))) ** 0.5
```

```python
# x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=2017)
# Train/Valid split
split = 25000
x_tr, y_tr, x_val, y_val = x_train[:split], y_train[:split], x_train[split:], y_train[split:]
```

```python
x_tr.fillna(-99999, inplace=True)
y_tr.fillna(-99999, inplace=True)
x_val.fillna(-99999, inplace=True)
y_val.fillna(-99999, inplace=True)
```

```python
xgtrain = xgb.DMatrix(x_tr, label=y_tr, missing='-99999.0')
xgval = xgb.DMatrix(x_val, label=y_val, missing='-99999.0')
```

```python
param = {'gamma':2, 'max_depth':2, 'eta':0.3, 'silent':0, 'objective':'reg:linear', 'subsample':0.8, 'feval': rmsle}
param['nthread'] = 3
plst = param.items()
# plst += [('eval_metric', 'mlogloss')] # Multiple evals can be handled in this way
# plst += [('feval', rmsle)]

evallist  = [(xgval,'eval'), (xgtrain,'train')]
```

```python
num_round = 500
model = xgb.train( plst, xgtrain, num_round, evallist, early_stopping_rounds=20, feval=rmsle_XGB)
```

```python
params = {}
params['objective'] = 'reg:linear'
params['eta'] = 0.02
params['silent'] = 1

watchlist = [(xgtrain, 'train'), (xgval, 'valid')]
clf = xgb.train(params, xgtrain, 800, watchlist, feval=rmsle_XGB, early_stopping_rounds=100)
```

```python
preds = clf.predict(xgtrain)
print 'Train:', rmsle(preds, y_tr, XGB=True)

preds = clf.predict(xgval)
print 'Validation:', rmsle(preds, y_val, XGB=True)
```

```python
p_test = clf.predict(xgb.DMatrix(x_test))

sub = pd.DataFrame()
sub['id'] = df_test['id'].values
sub['price_doc'] = p_test
sub.to_csv('submissions/xgb1.csv', index=False)
```
